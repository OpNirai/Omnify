Google_Ads data

```python
import pandas as pd
df=pd.read_csv('Omnify_Analyst.csv')
```

```python
df
```

```python
df['Week'] = pd.to_datetime(df['Week'])
df = df.sort_values(by='Week')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))

# Clicks over time
plt.subplot(2, 2, 1)
plt.plot(df['Week'], df['Clicks'], marker='o')
plt.title('Clicks over Time')
plt.xlabel('Week')
plt.ylabel('Clicks')
plt.xticks(rotation=45)
```

```python
# Cost ($) over time
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 2)
plt.plot(df['Week'], df['Cost'], marker='o', color='orange')
plt.title('Cost ($) over Time')
plt.xlabel('Week')
plt.ylabel('Cost ($)')
plt.xticks(rotation=45) 
```

```python
# Leads over time
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 3)
plt.plot(df['Week'], df['Leads'], marker='o', color='green')
plt.title('Leads over Time')
plt.xlabel('Week')
plt.ylabel('Leads')
plt.xticks(rotation=45) 
```

```python
df['Payment']=df['Payment'].str.replace('$','')
df['Payment']=df['Payment'].str.replace(',','')
df['Payment'] = pd.to_numeric(df['Payment'])
```

```python
# Payment ($) over time
plt.subplot(2, 2, 4)
plt.plot(df['Week'], df['Payment'], marker='o', color='red')
plt.title('Payment ($) over Time')
plt.xlabel('Week')
plt.ylabel('Payment ($)')
plt.xticks(rotation=45) 
```

Listing_site data

```python
import pandas as pd
df=pd.read_csv('Omnify_Analyst_1.csv')
```

```python
df
```

```python
df['Date of Report'] = pd.to_datetime(df['Date of Report'])
time_series_data = df.groupby('Date of Report').agg({
    'Clicks': 'sum',
    'MoneySpent': 'sum'
}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(time_series_data['Date of Report'], time_series_data['Clicks'], marker='o', label='Clicks')
plt.plot(time_series_data['Date of Report'], time_series_data['MoneySpent'], marker='o', label='MoneySpent')
plt.title('Clicks and MoneySpent Over Time')
plt.xlabel('Date of Report')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
df['Paid']=df['Paid'].str.replace('$','')
df['Paid']=df['Paid'].str.replace(',','')
df['Paid'] = pd.to_numeric(df['Paid'])
```

```python
import seaborn as sns
df_payment_made = df[df['Paid'] > 0]
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Date of Report', y='Paid', hue='Location', size='Paid', data=df_payment_made, palette='viridis', sizes=(50, 200))
plt.title('Payments by Location over Time')
plt.xlabel('Date of Report')
plt.ylabel('Payment ($)')
plt.xticks(rotation=45)
plt.legend(title='Location')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Date of Report', y='Paid', hue='Categories', size='Paid', data=df_payment_made, palette='viridis', sizes=(50, 200))
plt.title('Payments by Category over Time')
plt.xlabel('Date of Report')
plt.ylabel('Payment ($)')
plt.xticks(rotation=45)
plt.legend(title='Categories')
plt.grid(True)
plt.show()
```
