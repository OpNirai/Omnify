```python
import pandas as pd
df=pd.read_csv('Omnify_Analyst.csv')
```

```python
df
```

```python
df['Payment']=df['Payment'].str.replace('$','')
df['Payment']=df['Payment'].str.replace(',','')
df['Payment'] = pd.to_numeric(df['Payment'])
```

```python
df
```

```python
Amount_spent = df.pivot_table(index='Keywordtype', values='Cost', aggfunc='sum')
Amount_spent
```

```python
Amount_earned = df.pivot_table(index='Keywordtype', values='Payment', aggfunc='sum')
Amount_earned['profit']=Amount_earned['Payment']-Amount_spent['Cost']
Amount_earned
```

```python
print("The Keyword with maximum profit is",Amount_earned['profit'].index.max())
```
