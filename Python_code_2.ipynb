```python
import pandas as pd
df=pd.read_csv('Omnify_Analyst_1.csv')
```

```python
df['MoneySpent'] = pd.to_numeric(df['MoneySpent'])
```

```python
df
```

```python
Amount_spent = df.pivot_table(index='Channel', values='MoneySpent', aggfunc='sum')
Amount_spent
```

```python
df['Paid']=df['Paid'].str.replace('$','')
df['Paid']=df['Paid'].str.replace(',','')
df['Paid'] = pd.to_numeric(df['Paid'])
```

```python
Amount_earned = df.pivot_table(index='Channel', values='Paid', aggfunc='sum')
Amount_earned['profit']=Amount_earned['Paid']-Amount_spent['MoneySpent']
Amount_earned
```

```python
print("The Channel with maximum profit is",Amount_earned['profit'].index.max())
```
